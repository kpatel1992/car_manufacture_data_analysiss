## 1.1 Importing Libraries and Data

## 1.2 Merging Data Frames

```python
#Merge dataframe 1&2
merged_inner = pd.merge(left=df1, right=df2, how='inner', left_on='id', right_on='id')

#Create final data set of cars which contains all three files data with 27 attributes
cars = merged_inner.append(df3, ignore_index = True) 
cars
```

## 1.3 Data Cleaning

####    1.3.1 Checking Missing Values

```python
cars.isnull().sum()
#cars.isnull().values.any()
```

Normalised-losses and price are having null value NAN

##### Replace Null values 

```python
# Calculate Mean value of normalised losses
nlmean = cars['normalised-losses'].mean()
# Calculate Mean value of price
pmean = cars['price'].mean()

#Fill Null value with mean value
cars['normalised-losses'] = cars['normalised-losses'].fillna(nlmean)
cars['price'] = cars['price'].fillna(pmean)
```

#### 1.3.2 Check and remove duplicate Values

###### Check duplicate

```python
# check duplicate values
duplicateCar = cars[cars.duplicated(keep=False)]
print("All Duplicate Rows based on all columns are :")
duplicateCar
```

###### Remove duplicate

```python
# drop duplicate values
cars = cars.drop_duplicates()
```

#### 1.3.4 Typo Errors

```python
#cars['num-of-doors'].unique()
cars['num-of-doors'] = cars['num-of-doors'].replace(to_replace =4L, value ='four')
cars['num-of-doors'] = cars['num-of-doors'].replace(to_replace =2L, value ='two')
#cars['fuel-system'].unique()
cars['fuel-system'] = cars['fuel-system'].replace(to_replace ='Mpfi', value ='mpfi')
```

#### 1.3.5 Remove White Spaces

```python
# remove white space for string columns 
cars['fuel-type'] = cars['fuel-type'].replace(to_replace ='Diesel  ', value ='diesel')
cars['make'] = cars['make'].str.strip()
cars['fuel-type'] = cars['fuel-type'].str.strip()
cars['aspiration'] = cars['aspiration'].str.strip()
cars['num-of-doors'] = cars['num-of-doors'].str.strip()
cars['body-style'] = cars['body-style'].str.strip()
cars['drive-wheels'] = cars['drive-wheels'].str.strip()
cars['engine-location'] = cars['engine-location'].str.strip()
cars['engine-type'] = cars['engine-type'].str.strip()
cars['num-of-cylinders'] = cars['num-of-cylinders'].str.strip()
cars['fuel-system'] = cars['fuel-system'].str.strip()
```

#### 2.1.1 Highest No of vehicle Maker
